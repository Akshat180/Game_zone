```python
def checkpasswordstrength(password):
   
    sp='!@#$%^&*-=+/*?'            
    ac=dc=uc=lc=sc=st=0
    
    for i in password:
        if i.isalpha():
            ac+=1
            if i.islower():   lc+=1
            elif i.isupper(): lc+=1
        elif i.isdigit():     dc+=1
        elif i in sp:         sc+=1
            
    if len(password) > 8:    st+=1
    if ac>0 and dc >0:       st+=1
    if uc > 0 and lc>0:      st+=1
    if sc>0:                 st+=1
    if st>=3:                return'strong'
    else:                    return'weak'
```
